```python
import pandas as pd
import numpy as np
import yfinance as yf
from ta.momentum import RSIIndicator
from ta.trend import MACD, SMAIndicator, EMAIndicator
```

```python
stock_symbol = "AAPL"
data = yf.download(stock_symbol,period="5y", interval="1d")
data.dropna(inplace=True)
close_series = data["Close"].astype(float).values.flatten()
```

```python
data["SMA_10"] = SMAIndicator(close=pd.Series(close_series, index=data.index), window=10).sma_indicator()
data["SMA_20"] = SMAIndicator(close=pd.Series(close_series, index=data.index), window=20).sma_indicator()
data["EMA_10"] = EMAIndicator(close=pd.Series(close_series, index=data.index), window=10).ema_indicator()
data["EMA_20"] = EMAIndicator(close=pd.Series(close_series, index=data.index), window=20).ema_indicator()
```

```python
macd = MACD(close=pd.Series(close_series, index=data.index))
data["MACD"] = macd.macd()
data["MACD_Signal"] = macd.macd_signal()
```

```python
rsi = RSIIndicator(close=pd.Series(close_series, index=data.index), window=14)
data["RSI"] = rsi.rsi()
```

```python
data["Price_Change"] = data["Close"].pct_change()
data["Target"] = np.where(data["Price_Change"].shift(-1) > 0, 1, 0)
```

```python
data.dropna(inplace=True)
data.to_csv("../data/processed_stock_data.csv", index=False)
```
